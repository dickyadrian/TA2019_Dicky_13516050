```python
import cv2
import os
```

```python
vidcap = cv2.VideoCapture('video_tes.mp4')
fps = vidcap.get(cv2.CAP_PROP_FPS)

print('FPS is ' + str(fps))
```

```python
if not os.path.exists('data'):
    os.mkdir('data')
if not os.path.exists('result'):
    os.mkdir('result')
```

```python
import sys

success,image = vidcap.read()
frame = 1
group = 0
FRAMES_PATH = os.path.join('data', 'frames_')
RESULT_PATH = 'result'
if not os.path.exists(FRAMES_PATH+str(group)):
    os.mkdir(FRAMES_PATH+str(group))
while success:
    if (frame % 2 != 0):
        #write to frames
        cv2.imwrite(os.path.join(FRAMES_PATH + str(group), 'frame' + str(frame) + '.jpg'), image)
        if (frame != 1):
            group += 1
            os.mkdir(FRAMES_PATH+str(group))
            if (frame % 2 != 0):
                cv2.imwrite(os.path.join(FRAMES_PATH + str(group), 'frame' + str(frame) + '.jpg'), image)
    else: 
        #write to result
        cv2.imwrite(os.path.join(RESULT_PATH, 'frame' + str(frame) + '.jpg'), image)
    success,image = vidcap.read()
    frame += 1
```
